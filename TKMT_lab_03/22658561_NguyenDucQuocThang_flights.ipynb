```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file
data = pd.read_csv('flights.csv')

# Tính tổng số lượng hành khách theo năm
passengers_by_year = data.groupby(data['year'])['passengers'].sum()
```

```python
# Vẽ biểu đồ cột số lượng hành khách theo năm
passengers_by_year.plot(kind='bar', edgecolor ='k',figsize=(10, 6))
plt.xlabel('Năm')
plt.ylabel('Số lượng hành khách')
plt.title('Số lượng hành khách qua đi máy bay theo năm')
plt.show()

# Tính tổng số lượng hành khách theo tháng
passengers_by_month = data.groupby(data['month'])['passengers'].sum()
```

```python
# Vẽ biểu đồ đường số lượng hành khách theo tháng
passengers_by_month.plot(kind='line', marker='o', figsize=(10, 6))
plt.xlabel('Tháng')
plt.ylabel('Số lượng hành khách')
plt.title('Số lượng hành khách qua đi máy bay theo tháng')
plt.xticks(range(len(passengers_by_month)), passengers_by_month.index)
plt.show()
```
