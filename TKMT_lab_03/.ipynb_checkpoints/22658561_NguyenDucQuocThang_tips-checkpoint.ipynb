```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu từ file
data = pd.read_csv('tips.csv')

# Tổng quan về dữ liệu
print("Thông tin dữ liệu:")
print(data.info())
```

```python
# Thống kê mô tả các đại lượng
print("\nThống kê mô tả:")
print(data.describe())
```

```python
# Phân tích tiền tip theo giới tính
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='tip', data=data)
plt.xlabel('Giới tính')
plt.ylabel('Tiền tip (USD)')
plt.title('Phân tích tiền tip theo giới tính')
plt.show()
```

```python
# Phân tích tiền tip theo việc hút thuốc
plt.figure(figsize=(8, 6))
sns.boxplot(x='smoker', y='tip', data=data)
plt.xlabel('Hút thuốc')
plt.ylabel('Tiền tip (USD)')
plt.title('Phân tích tiền tip theo việc hút thuốc')
plt.show()
```

```python
# Phân tích tiền tip theo ngày trong tuần
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='tip', data=data)
plt.xlabel('Ngày trong tuần')
plt.ylabel('Tiền tip (USD)')
plt.title('Phân tích tiền tip theo ngày trong tuần')
plt.show()
```

```python
# Phân tích tiền tip theo số lượng người
plt.figure(figsize=(8, 6))
sns.boxplot(x='size', y='tip', data=data)
plt.xlabel('Số lượng người')
plt.ylabel('Tiền tip (USD)')
plt.title('Phân tích tiền tip theo số lượng người')
plt.show()
```
