```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file
data = pd.read_csv('babies.txt', sep='\s+')

# Tách dữ liệu thành hai nhóm: nhóm thai phụ hút thuốc và nhóm thai phụ không hút thuốc
smoke_group = data[data['smoke'] == 1]
non_smoke_group = data[data['smoke'] == 0]

# Tính giá trị trung bình và độ lệch chuẩn của trọng lượng (bwt) trong từng nhóm
smoke_mean = smoke_group['bwt'].mean()
smoke_std = smoke_group['bwt'].std()

non_smoke_mean = non_smoke_group['bwt'].mean()
non_smoke_std = non_smoke_group['bwt'].std()

# Hiển thị kết quả
print("Nhóm thai phụ hút thuốc:")
print(f"Giá trị trung bình trọng lượng: {smoke_mean}")
print(f"Độ lệch chuẩn trọng lượng: {smoke_std}")

print("\nNhóm thai phụ không hút thuốc:")
print(f"Giá trị trung bình trọng lượng: {non_smoke_mean}")
print(f"Độ lệch chuẩn trọng lượng: {non_smoke_std}")

# Vẽ biểu đồ histogram so sánh trọng lượng giữa nhóm thai phụ hút thuốc và không hút thuốc
plt.hist(smoke_group['bwt'], bins=20, alpha=0.5, label='Hút thuốc')
plt.hist(non_smoke_group['bwt'], bins=20, alpha=0.5, label='Không hút thuốc')
plt.xlabel('Trọng lượng (ounces)')
plt.ylabel('Số lượng')
plt.title('Phân bố trọng lượng sơ sinh')
plt.legend()
plt.show()
```
