**Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import norm
from scipy.stats import t
# a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. 
# Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.
n, p = 10, 0.25
cau_dung = binom.rvs(n, p,size=10)
print("Số câu đúng của sinh viên:", cau_dung)
# b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm
p_4_diem = binom.pmf(4, n, p)
print("Xác suất để sinh viên được 4 điểm:", p_4_diem)
# c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)
E_X = binom.mean(n, p)
print("Giá trị trung bình của X:", E_X)
Var_X = binom.var(n, p)
print("Phương sai của X:", Var_X)
```

**Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
lambda_ = 3  # Lambda là giá trị kỳ vọng số cuộc gọi trong một phút

# Tính xác suất nhận được 1 cuộc gọi
prob_1_call = poisson.pmf(1, lambda_)
print(f"Xác suất nhận được 1 cuộc gọi: {prob_1_call}")

# Tính xác suất nhận được 2 cuộc gọi
prob_2_calls = poisson.pmf(2, lambda_)
print(f"Xác suất nhận được 2 cuộc gọi: {prob_2_calls}")

# Tính xác suất nhận được 3 cuộc gọi
prob_3_calls = poisson.pmf(3, lambda_)
print(f"Xác suất nhận được 3 cuộc gọi: {prob_3_calls}")
```

**Bài 03:** Trọng lượng (đơn vị gam) của một loại trái cây có phân phối chuẩn với µ = 500 (gam) gam và $σ^2 = 16 (gam^2)$. Trái cây thu hoạch được phân loại theo trọng lượng như sau:

Loại 1: trên 505 gam

Loại 2: từ 495 – 505 gam

Loại 3: dười 495 gam

a. Hãy tính tỷ lệ của mỗi loại.
b. Bạn hãy mô phỏng lấy mẫu  20 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
c. Bạn hãy mô phỏng lấy 100 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?

```python
import numpy as np
import matplotlib.pyplot as plt

# Thông số phân phối chuẩn
mu = 500
sigma_squared = 16

# a. Tính tỷ lệ của mỗi loại trái cây
prob_type1 = 1 - norm.cdf(505, mu, np.sqrt(sigma_squared))
prob_type2 = norm.cdf(505, mu, np.sqrt(sigma_squared)) - norm.cdf(495, mu, np.sqrt(sigma_squared))
prob_type3 = norm.cdf(495, mu, np.sqrt(sigma_squared))

print(f"Tỷ lệ loại 1: {prob_type1}")
print(f"Tỷ lệ loại 2: {prob_type2}")
print(f"Tỷ lệ loại 3: {prob_type3}")

# b. Mô phỏng lấy mẫu 20 trái cây
sample_size = 20
sample = np.random.normal(mu, np.sqrt(sigma_squared), sample_size)

# Vẽ histogram trọng lượng trái cây
plt.hist(sample, bins=10, edgecolor='black')
plt.xlabel('Trọng lượng (gam)')
plt.ylabel('Số lượng')
plt.title('Histogram trọng lượng trái cây (mẫu 20)')
plt.show()

# Tính giá trị trung bình và độ lệch chuẩn từ mẫu
sample_mean = np.mean(sample)
sample_std = np.std(sample)

print(f"Giá trị trung bình từ mẫu 20: {sample_mean}")
print(f"Độ lệch chuẩn từ mẫu 20: {sample_std}")

# c. Mô phỏng lấy mẫu 100 trái cây
sample_size = 100
sample = np.random.normal(mu, np.sqrt(sigma_squared), sample_size)

# Vẽ histogram trọng lượng trái cây
plt.hist(sample, bins=10, edgecolor='black')
plt.xlabel('Trọng lượng (gam)')
plt.ylabel('Số lượng')
plt.title('Histogram trọng lượng trái cây (mẫu 100)')
plt.show()

# Tính giá trị trung bình và độ lệch chuẩn từ mẫu
sample_mean = np.mean(sample)
sample_std = np.std(sample)

print(f"Giá trị trung bình từ mẫu 100: {sample_mean}")
print(f"Độ lệch chuẩn từ mẫu 100: {sample_std}")

# d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?

# Từ hai cách lấy mẫu ở câu b và câu c, ta có thể nhận thấy một số điểm sau:

# Kích thước mẫu: Mẫu 100 trái cây có kích thước lớn hơn mẫu 20 trái cây. 
# Khi kích thước mẫu tăng lên, chúng ta có một tập dữ liệu lớn hơn để tính toán các thống kê và 
# ước lượng. Điều này có thể làm cho kết quả đạt được từ mẫu 100 trái cây có tính ổn định và 
# chính xác hơn so với mẫu 20 trái cây.

# Độ lệch chuẩn: Độ lệch chuẩn từ mẫu 100 trái cây có thể xấp xỉ với độ lệch chuẩn của phân phối chuẩn
# ban đầu hơn so với độ lệch chuẩn từ mẫu 20 trái cây. Khi kích thước mẫu tăng lên, độ lệch chuẩn từ 
# mẫu sẽ càng tiến gần đến giá trị độ lệch chuẩn thực tế của phân phối.

# Sự phân phối: Khi kích thước mẫu tăng lên, biểu đồ histogram của mẫu 100 trái cây có khả năng phản 
# ánh chính xác hơn hình dạng phân phối chuẩn so với biểu đồ của mẫu 20 trái cây.

# # Tổng quan, việc lấy mẫu lớn hơn (mẫu 100 trái cây) thường mang lại kết quả ước lượng tốt hơn và 
# gần với giá trị thực tế của phân phối chuẩn hơn. Điều này cho thấy tầm quan trọng của kích thước mẫu 
# trong việc đảm bảo tính chính xác của ước lượng thống kê.
```
