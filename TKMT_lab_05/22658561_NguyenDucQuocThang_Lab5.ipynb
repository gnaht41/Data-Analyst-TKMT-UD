### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Tạo mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên
np.random.seed(0)
POP = np.random.uniform(120, 200, 10000)

# a. Tính chiều cao trung bình và độ lệch chuẩn của quần thể
mean_height = np.mean(POP)
std_dev_height = np.std(POP)

print("a. Chiều cao trung bình của quần thể:", mean_height)
print("   Độ lệch chuẩn về chiều cao của quần thể:", std_dev_height)

# b. Tính tỷ lệ người cao trong quần thể
threshold = 180
tall_ratio = np.mean(POP >= threshold)

print("\nb. Tỷ lệ người cao trong quần thể:", tall_ratio)

# c. Vẽ histogram về chiều cao của quần thể
plt.figure(figsize=(10, 6))
plt.hist(POP, bins=50, density=True, alpha=0.7)
plt.xlabel('Chiều cao')
plt.ylabel('Tần suất')
plt.title('Histogram của quần thể')
plt.show()

# d. Vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn
x = np.linspace(mean_height - 4 * std_dev_height, mean_height + 4 * std_dev_height, 100)
pdf = stats.norm.pdf(x, mean_height, std_dev_height)

plt.figure(figsize=(10, 6))
plt.hist(POP, bins=50, density=True, alpha=0.7, label='Quần thể')
plt.plot(x, pdf, 'r', label='Phân phối chuẩn')
plt.xlabel('Chiều cao')
plt.ylabel('Mật độ')
plt.title('So sánh Histogram và Phân phối chuẩn')
plt.legend()
plt.show()
```

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo danh sách rỗng để chứa trung bình mẫu
sample_means_bai2 = []
sample_means_bai3 = []

# Thực hiện bài 2 và tính trung bình mẫu 100 lần
for _ in range(100):
    sample = np.random.choice(POP, size=20, replace=False)
    mean_sample = np.mean(sample)
    sample_means_bai2.append(mean_sample)

# Thực hiện bài 3 và tính trung bình mẫu 100 lần
for _ in range(100):
    sample = np.random.choice(POP, size=100, replace=False)
    mean_sample = np.mean(sample)
    sample_means_bai3.append(mean_sample)

# a. Vẽ histogram cho 100 trung bình mẫu của bài 2
plt.figure(figsize=(10, 6))
plt.hist(sample_means_bai2, bins=10, density=True, alpha=0.7)
plt.xlabel('Trung bình mẫu')
plt.ylabel('Tần suất')
plt.title('Histogram của trung bình mẫu (Bài 2)')
plt.show()

# b. Vẽ histogram cho 100 trung bình mẫu của bài 3
plt.figure(figsize=(10, 6))
plt.hist(sample_means_bai3, bins=10, density=True, alpha=0.7)
plt.xlabel('Trung bình mẫu')
plt.ylabel('Tần suất')
plt.title('Histogram của trung bình mẫu (Bài 3)')
plt.show()

# c. Vẽ histogram của cả hai bài trong cùng một hình để so sánh
plt.figure(figsize=(10, 6))
plt.hist(sample_means_bai2, bins=10, density=True, alpha=0.7, label='Bài 2')
plt.hist(sample_means_bai3, bins=10, density=True, alpha=0.7, label='Bài 3')
plt.xlabel('Trung bình mẫu')
plt.ylabel('Tần suất')
plt.title('So sánh Histogram của trung bình mẫu (Bài 2 và Bài 3)')
plt.legend()
plt.show()
```

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo danh sách rỗng để chứa phương sai của các mẫu
sample_variances = []

# Mô phỏng 1000 lần lấy mẫu có kích thước 100 và tính phương sai
for _ in range(1000):
    sample = np.random.choice(POP, size=100, replace=False)
    sample_variance = np.var(sample)
    sample_variances.append(sample_variance)

# Vẽ histogram đại lượng phương sai của các mẫu
plt.figure(figsize=(10, 6))
plt.hist(sample_variances, bins=20, density=True, alpha=0.7)
plt.xlabel('Phương sai')
plt.ylabel('Tần suất')
plt.title('Histogram đại lượng phương sai của các mẫu')
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
import numpy as np
import scipy.stats as stats

data = [2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8]
confidence = 0.95

# a) Khi biết σ = 0.3
sigma_known = 0.3
mean = np.mean(data)
n = len(data)
z = stats.norm.ppf((1 + confidence) / 2)
margin_of_error = z * (sigma_known / np.sqrt(n))
confidence_interval_a = (mean - margin_of_error, mean + margin_of_error)

# b) Khi không biết σ
std_sample = np.std(data, ddof=1)
t = stats.t.ppf((1 + confidence) / 2, df=n - 1)
margin_of_error = t * (std_sample / np.sqrt(n))
confidence_interval_b = (mean - margin_of_error, mean + margin_of_error)

print("Khoảng tin cậy (biết σ):", confidence_interval_a)
print("Khoảng tin cậy (không biết σ):", confidence_interval_b)
```
