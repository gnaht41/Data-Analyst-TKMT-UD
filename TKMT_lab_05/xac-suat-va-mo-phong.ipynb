## 1. Biến ngẫu nhiên rời rạc

Giả sử X là biến ngẫu nhiên rời rạc

a. Hãy cho biết ý nghĩa câu lệnh sau: `binom.pmf(4, 10, 0.6)`

b. Giả sử X có phân phối nhị thức `X ~ B(10, 0.6)`. Hãy tính các giá trị sau:  
$P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

c. Hãy cho biết ý nghĩa câu lệnh sau: `poisson.pmf(4, 3)`

d. Giả sử X có phân phối Poisson `X ~ P(3)`. Hãy tính các giá trị sau:  
$P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

```python
import scipy.stats as stats
# a. Ý nghĩa của câu lệnh binom.pmf(4, 10, 0.6)
#    Tính xác suất hàm khối xác suất (PMF) của phân phối nhị thức
#    với n = 10 (số lần thử), p = 0.6 (xác suất thành công)
#    để X = 4 (số lần thành công mong muốn)
binomial_pmf = stats.binom.pmf(4, 10, 0.6)
print("a. Ý nghĩa câu lệnh binom.pmf(4, 10, 0.6):", binomial_pmf)

# b. Tính các giá trị cho phân phối nhị thức X ~ B(10, 0.6)
binomial_cdf_5 = stats.binom.cdf(5, 10, 0.6)  # P(X ≤ 5)
binomial_cdf_4 = stats.binom.cdf(4, 10, 0.6)  # P(X ≤ 4)
binomial_sf_3 = stats.binom.sf(3, 10, 0.6)    # P(X ≥ 4)
binomial_pmf_5 = stats.binom.pmf(5, 10, 0.6)   # P(X = 5)
print("b. Các giá trị cho phân phối nhị thức X ~ B(10, 0.6):")
print("   P(X ≤ 5):", binomial_cdf_5)
print("   P(X ≤ 4):", binomial_cdf_4)
print("   P(X ≥ 4):", binomial_sf_3)
print("   P(X = 5):", binomial_pmf_5)

# c. Ý nghĩa của câu lệnh poisson.pmf(4, 3)
#    Tính hàm khối xác suất (PMF) của phân phối Poisson
#    với λ = 3 (tham số trung bình)
#    để X = 4 (số lần xảy ra sự kiện mong muốn)
poisson_pmf = stats.poisson.pmf(4, 3)
print("c. Ý nghĩa câu lệnh poisson.pmf(4, 3):", poisson_pmf)

# d. Tính các giá trị cho phân phối Poisson X ~ P(3)
poisson_cdf_5 = stats.poisson.cdf(5, 3)  # P(X ≤ 5)
poisson_cdf_4 = stats.poisson.cdf(4, 3)  # P(X ≤ 4)
poisson_sf_3 = stats.poisson.sf(3, 3)    # P(X ≥ 4)
poisson_pmf_5 = stats.poisson.pmf(5, 3)  # P(X = 5)
print("d. Các giá trị cho phân phối Poisson X ~ P(3):")
print("   P(X ≤ 5):", poisson_cdf_5)
print("   P(X ≤ 4):", poisson_cdf_4)
print("   P(X ≥ 4):", poisson_sf_3)
print("   P(X = 5):", poisson_pmf_5)
```

# 2. Tỷ lệ đậu

Một môn học X được tổ chức mỗi học kỳ có tỷ lệ thí sinh học qua môn trung bình là 80%, và mỗi học kỳ có 250 sinh viên đăng ký dự thi. Tính giá trị kỳ vọng và độ lệch chuẩn của số lượng thí sinh không vượt qua môn học?

```python
import math

# Tỉ lệ thí sinh học qua môn trung bình
p = 0.8

# Số lượng sinh viên đăng ký dự thi
n = 250

# Tính giá trị kỳ vọng
expected_value = n * p

# Tính độ lệch chuẩn
standard_deviation = math.sqrt(n * p * (1 - p))

print("Giá trị kỳ vọng (Expected value) của số lượng thí sinh không vượt qua môn học:", expected_value)
print("Độ lệch chuẩn (Standard deviation) của số lượng thí sinh không vượt qua môn học:", standard_deviation)
```

## 3. Tổng đài

Một tổng đài trung bình trong một giờ nhận được 180 cuộc gọi một cách ngẫu nhiên. Nếu trong 5 phút có từ hơn 20 cuộc gọi trở lên, thì tổng đài sẽ quá tải và người gọi đến sẽ phải chờ đợi. Giả sử số cuộc gọi nhận được trong 5 phút của tổng đài có phân phối Poisson. Tính:

a. Xác suất nhận được nhiều hơn 19 cuộc gọi mỗi 5 phút.

b. Trong 5 phút tổng đài có thể nhận được nhiều nhất bao nhiêu cuộc gọi để xác suất các cuộc gọi đều được xử lý ít nhất là 99%

```python
import scipy.stats as stats

# Trung bình số cuộc gọi nhận được trong một giờ
lambda_ = 180 / 60  # Mỗi 5 phút

# a. Xác suất nhận được nhiều hơn 19 cuộc gọi mỗi 5 phút
probability_a = 1 - stats.poisson.cdf(19, lambda_)
print("a. Xác suất nhận được nhiều hơn 19 cuộc gọi mỗi 5 phút:", probability_a)

# b. Tìm số cuộc gọi tối đa để xác suất các cuộc gọi đều được xử lý ít nhất là 99%
threshold_probability = 0.99
max_calls = stats.poisson.ppf(threshold_probability, lambda_)
print("b. Số cuộc gọi tối đa để xác suất các cuộc gọi đều được xử lý ít nhất là 99%:", max_calls)
```

## 4. Phân phối liên tục

a. Cho biết ý nghĩa của các câu lệnh sau:

- `norm.cdf(2)`

- `norm.cdf(2, 1, 1)`

- `norm.cdf(2, 1, 2)`

b. Cho biết kết quả của câu lệnh sau: `norm.ppf(norm.cdf(2))`

c. Tính các giá trị sau và vẽ hình minh họa cho các giá trị tính được:

- `norm.ppf(0.975)`

- `norm.ppf(0.975, 1, 1)`

- `norm.ppf(0.975, 1, 2)`

d. Giả sử biến ngẫu nhiên X có phân phối chuẩn với kỳ vọng là 24 và phương sai là 16. Hãy tính các giá trị sau:

- $P(X \leq 20)$
- $P(X \ge 29.5)$
- $P(X = 23.8)$

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# a. Ý nghĩa của các câu lệnh
result_a1 = stats.norm.cdf(2)
result_a2 = stats.norm.cdf(2, 1, 1)
result_a3 = stats.norm.cdf(2, 1, 2)

print("a. Kết quả các câu lệnh:")
print("norm.cdf(2):", result_a1)
print("norm.cdf(2, 1, 1):", result_a2)
print("norm.cdf(2, 1, 2):", result_a3)

# b. Kết quả của câu lệnh
result_b = stats.norm.ppf(stats.norm.cdf(2))
print("\nb. Kết quả của câu lệnh norm.ppf(norm.cdf(2)):", result_b)

# c. Tính giá trị và vẽ hình minh họa
result_c1 = stats.norm.ppf(0.975)
result_c2 = stats.norm.ppf(0.975, 1, 1)
result_c3 = stats.norm.ppf(0.975, 1, 2)

print("\nc. Kết quả các câu lệnh:")
print("norm.ppf(0.975):", result_c1)
print("norm.ppf(0.975, 1, 1):", result_c2)
print("norm.ppf(0.975, 1, 2):", result_c3)

x = np.linspace(-5, 5, 1000)
y1 = stats.norm.pdf(x)
y2 = stats.norm.pdf(x, 1, 1)
y3 = stats.norm.pdf(x, 1, 2)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="Standard Normal Distribution")
plt.plot(x, y2, label="Mean=1, SD=1")
plt.plot(x, y3, label="Mean=1, SD=2")
plt.legend()
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distributions')
plt.show()

# d. Tính giá trị
mean = 24
std_dev = 4

result_d1 = stats.norm.cdf(20, mean, std_dev)
result_d2 = 1 - stats.norm.cdf(29.5, mean, std_dev)
result_d3 = stats.norm.pdf(23.8, mean, std_dev)

print("\nd. Kết quả các câu lệnh:")
print("P(X <= 20):", result_d1)
print("P(X >= 29.5):", result_d2)
print("P(X = 23.8):", result_d3)
```

## 5. Độ dài chi tiết máy

Một công ty sản xuất làm ra các chi tiết máy với độ dài các chi tiết có phân phối chuẩn với kỳ vọng là 3000 mm và độ lệch chuẩn là 3 mm. Yêu cầu cho các chi tiết máy này là phải có chiều dài từ 2993 mm đến 3007 mm. Tính tỷ lệ sản phẩm lỗi của công ty

```python
import scipy.stats as stats

mean = 3000  # Kỳ vọng
std_dev = 3  # Độ lệch chuẩn

lower_bound = 2993  # Giới hạn dưới
upper_bound = 3007  # Giới hạn trên

# Tính tỷ lệ sản phẩm lỗi
error_rate = 1 - stats.norm.cdf(upper_bound, mean, std_dev) + stats.norm.cdf(lower_bound, mean, std_dev)

print("Tỷ lệ sản phẩm lỗi của công ty:", error_rate)
```

## 6. Mô phỏng phân phối Chuẩn

Giả sử điểm thi TOEIC là một biến ngẫu nhiên có phân phối chuẩn với kỳ vọng là 500, độ lệch chuẩn là 8. Bạn hãy:

a. Mô phỏng bằng cách phát sinh ngẫu nhiên điểm thi TOIEC của một nhóm gồm 50 người.

b. Tính điểm trung bình và độ lệch chuẩn về điểm của nhóm.

c. Tìm miền giá trị, và miền phân vị (IQR) của nhóm

d. Cho biết tỷ lệ đạt trên 450 điểm của nhóm. 

e. Vẽ đồ thị histogram của nhóm và so sánh với đồ thị hàm mật độ của phân phối chuẩn với kỳ vọng là 500 và độ lệch chuẩn là 8.

f. Thay vì mô phỏng 50 người, hãy mô phỏng nhóm với 100 người thi. Bạn có nhận xét gì?

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# a. Mô phỏng điểm thi TOIEC của một nhóm gồm 50 người
np.random.seed(0)
mean = 500
std_dev = 8
num_people = 50
scores = np.random.normal(mean, std_dev, num_people)

# b. Tính điểm trung bình và độ lệch chuẩn của nhóm
mean_group = np.mean(scores)
std_dev_group = np.std(scores)

print("b. Điểm trung bình của nhóm:", mean_group)
print("   Độ lệch chuẩn về điểm của nhóm:", std_dev_group)

# c. Tìm miền giá trị và miền phân vị của nhóm
min_value = np.min(scores)
max_value = np.max(scores)
iqr = stats.iqr(scores)

print("\nc. Miền giá trị của nhóm:", min_value, "-", max_value)
print("   Miền phân vị (IQR) của nhóm:", iqr)

# d. Tính tỷ lệ đạt trên 450 điểm của nhóm
pass_threshold = 450
pass_rate = np.mean(scores > pass_threshold)

print("\nd. Tỷ lệ đạt trên 450 điểm của nhóm:", pass_rate)

# e. Vẽ đồ thị histogram và so sánh với đồ thị hàm mật độ của phân phối chuẩn
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 100)
pdf = stats.norm.pdf(x, mean, std_dev)

plt.figure(figsize=(10, 6))
plt.hist(scores, bins=10, density=True, alpha=0.7, label='Group Scores')
plt.plot(x, pdf, 'r', label='Normal Distribution')
plt.xlabel('Scores')
plt.ylabel('Density')
plt.title('Histogram of Group Scores')
plt.legend()
plt.show()

# f. Mô phỏng nhóm với 100 người thi
num_people_new = 100
scores_new = np.random.normal(mean, std_dev, num_people_new)

mean_group_new = np.mean(scores_new)
std_dev_group_new = np.std(scores_new)

print("\nf. Với nhóm 100 người:")
print("   Điểm trung bình của nhóm:", mean_group_new)
print("   Độ lệch chuẩn về điểm của nhóm:", std_dev_group_new)
```
